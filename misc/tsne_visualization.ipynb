```python
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

reference: https://scipy-lectures.org/packages/scikit-learn/auto_examples/plot_tsne.html

```python
emb = np.loadtxt('test-embs-vecs.tsv', delimiter='\t')
# skip label header
words, suffixes = np.loadtxt('test-embs-meta.tsv', delimiter='\t', dtype=str, skiprows=1, unpack=True)
```

```python
tsne = TSNE(perplexity=8, n_components=3, init='pca')
emb_2d = tsne.fit_transform(emb)
```

```python
colors = {'ing': 'r', 'tion': 'g', 'ly': 'b'}
plt.figure(figsize=(10, 7))
for idx in range(emb.shape[0]):
    plt.scatter(emb_2d[idx, 0], emb_2d[idx, 1], color=colors[suffixes[idx]])
    plt.annotate(words[idx], (emb_2d[idx, 0], emb_2d[idx, 1]), color=colors[suffixes[idx]])
plt.savefig('tsne.png')
plt.show()
```
